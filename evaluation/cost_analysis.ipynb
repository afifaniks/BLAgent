```python
import json
```

```python
# phase1_agentic_cost_path = "../results/cost_analysis/gpt_oss_phase1.json"
phase1_agentic_cost_path = "../results/cost_analysis/claude_4_6_phase1.json"

with open(phase1_agentic_cost_path, "r") as f:
    agentic_cost_data = json.load(f)

if "claude" in phase1_agentic_cost_path:
    input_cost_per_token = 0.000003
    output_cost_per_token = 0.000015

# https://novita.ai/models/model-detail/openai-gpt-oss-120b
elif "gpt" in phase1_agentic_cost_path:
    input_cost_per_token = 0.05/1000000
    output_cost_per_token = 0.25/1000000
```

```python
prompt_tokens = 0
completion_tokens = 0

for entry in agentic_cost_data:
    token_usage = entry.get("token_usage", {})
    prompt_tokens += token_usage.get("prompt_tokens", 0)
    completion_tokens += token_usage.get("completion_touokens", 0)

total_tokens = prompt_tokens + completion_tokens

avg_prompt_tokens_agent = prompt_tokens / len(agentic_cost_data)
avg_completion_tokens_agent = completion_tokens / len(agentic_cost_data)
avg_total_tokens_agent = total_tokens / len(agentic_cost_data)

print(f"Average Prompt Tokens: {avg_prompt_tokens_agent:.2f}")
print(f"Average Completion Tokens: {avg_completion_tokens_agent:.2f}")
print(f"Total Average Tokens: {avg_total_tokens_agent:.2f}")

print(f"Input Token Cost: ${avg_prompt_tokens_agent * input_cost_per_token:.4f}")
print(f"Output Token Cost: ${avg_completion_tokens_agent * output_cost_per_token:.4f}")
print(f"Total Cost: ${avg_prompt_tokens_agent * input_cost_per_token + avg_completion_tokens_agent * output_cost_per_token:.4f}")
```

```python
# reranker_data_path = "../results/cost_analysis/gpt_oss_phase2.json"
reranker_data_path = "../results/cost_analysis/claude_4_6_phase2.json"

with open(reranker_data_path, "r") as f:
    reranker_data = json.load(f)
```

```python
prompt_tokens = 0
completion_tokens = 0

for entry in reranker_data:
    token_usage = entry.get("token_usage", {})
    prompt_tokens += token_usage.get("prompt_tokens", 0)
    completion_tokens += token_usage.get("output_tokens", 0)

total_tokens = prompt_tokens + completion_tokens

avg_prompt_tokens_rag = prompt_tokens / len(reranker_data)
avg_completion_tokens_rag = completion_tokens / len(reranker_data)
avg_total_tokens_rag = total_tokens / len(reranker_data)

print(f"Average Prompt Tokens: {avg_prompt_tokens_rag:.2f}")
print(f"Average Completion Tokens: {avg_completion_tokens_rag:.2f}")
print(f"Total Average Tokens: {avg_total_tokens_rag:.2f}")
print(f"Input Token Cost: ${avg_prompt_tokens_rag * input_cost_per_token:.4f}")
print(f"Output Token Cost: ${avg_completion_tokens_rag * output_cost_per_token:.4f}")
print(f"Total Cost: ${avg_prompt_tokens_rag * input_cost_per_token + avg_completion_tokens_rag * output_cost_per_token:.4f}")
```

```python
total_input_in_both = avg_prompt_tokens_agent + avg_prompt_tokens_rag
total_output_in_both = avg_completion_tokens_agent + avg_completion_tokens_rag
total_cost_in_both = total_input_in_both * input_cost_per_token + total_output_in_both * output_cost_per_token
print(f"Total Input Tokens in Both: {total_input_in_both:.2f}")
print(f"Total Output Tokens in Both: {total_output_in_both:.2f}")
print(f"Total Cost in Both: ${total_cost_in_both:.4f}")
```
